```python
#вспомогательные функции
from typing import List, Tuple, Dict
def is_palindrome(x: int) -> bool:
    s = str(x)
    return s == s[::-1]

def is_prime(n: int) -> bool:
    return n >= 2 and all(n % i != 0 for i in range(2,int(n**0.5)+1))

def permutations(num: int) -> List[int]:
    s = str(num)
    perms = []
    for i in range(len(s)):
        perm = int(s[i:] + s[:i])
        perms.append(perm)
    return perms
```

```python
#Задание 1

def palindromic_squares_and_circular_primes() -> Tuple[List[int], List[int]]:
    palindromic_squares = []
    circular_primes = []

    for a in range(10**5):
        if is_palindrome(a) and is_palindrome(a ** 2):
            palindromic_squares.append(a)

    for p in range(2, 10**6):
        if is_prime(p):
            perms = permutations(p)
            if all(is_prime(x) for x in perms):
                circular_primes.append(p)
    return palindromic_squares, circular_primes

result = palindromic_squares_and_circular_primes()
result1_1, result1_2 = result
```

```python
#Задание 3
def primes_with_two_digits() -> Dict[str, List[int]]:
    """
    Возвращает словарь вида:
    {
    '13': [список первых 100 простых из {1,3}],
    '15': [список первых 100 простых из {1,5}],
    '17': [список первых 100 простых из {1,7}],
    '19': [список первых 100 простых из {1,9}]
    }
    """
    answer = {'13': [], '15': [], '17': [], '19': []}
    digits_list = answer.keys()
    for digits in digits_list:
      current = [0]
      while len(answer[digits]) < 100:
        new_list = [10 * old_num + int(digit) for old_num in current for digit in digits]
        for num in new_list:
          if is_prime(num) and len(answer[digits]) < 100:
            answer[digits].append(num)
        current = new_list
    return answer
result3 = primes_with_two_digits()
```

```python
#Задание 4:
def twin_primes_analysis(limit_pairs: int = 1000) -> Tuple[List[Tuple[int, int]],List[float]]:
  """
  Возвращает:
  - список первых `limit_pairs` пар близнецов (p, p+2);
  - список значений отношения pi_2(n) / pi(n) для n,
   соответствующих последним элементам каждой пары,
  где pi_2(n) — количество пар близнецов <= n, pi(n) — количество простых <= n.
  """
  total_prime = 1
  prime_pairs = []
  n = 2
  frequency = []
  while len(prime_pairs) < limit_pairs:
    if is_prime(n):
      total_prime+=1
      if is_prime(n+2):
        pair = (n,n+2)
        prime_pairs.append(pair)
        frequency.append(len(prime_pairs)/total_prime)
    n+=1
  return prime_pairs, frequency


result_4 =twin_primes_analysis(1000)
```

```python
#задача 5_1
from typing import Dict
from sympy import factorint
from math import factorial

def factorial_plus_one_factors() -> Dict[int, Dict[int, int]]:
    """
    Возвращает словарь вида:
    { n: {простой_делитель: степень, ...}, ... }
    для n от 2 до 50, где ключ — n,
    значение — разложение n! + 1 на простые множители.
    """
    result = {}
    for n in range(2, 51):
        val = factorial(n) + 1
        factors = factorint(val)  # sympy факторизация — быстрая и эффективная
        result[n] = factors
    return result

result_5_1 = factorial_plus_one_factors()
```

```python
#задача 5_2
def large_prime_factors_cases(threshold: int = 10**6) -> Dict[int, int]:
    """
    Для n от 2 до 50 возвращает словарь {n: самый_большой_простой_множитель},
    если среди простых делителей n! + 1 есть множитель больше threshold.
    """
    cases = {}
    for n in range(2, 51):
        val = factorial(n) + 1
        factors = factorint(val)
        large_primes = [p for p in factors if p > threshold]
        if large_primes:
            cases[n] = max(large_primes)
    return cases

result_5_2 = large_prime_factors_cases()
```

```python
#задача 6
import time
from typing import List, Dict
from math import gcd
from sympy import factorint
from sympy.ntheory import totient

def euler_phi_direct(n: int) -> int:
    """Вычисляет φ(n) прямым перебором."""
    count = 0
    for k in range(1, n + 1):
        if gcd(k, n) == 1:
            count += 1
    return count

def euler_phi_factor(n: int) -> int:
    """Вычисляет φ(n) через разложение на простые множители."""
    factors = factorint(n)
    result = n
    for p in factors:
        result = result * (1 - 1/p)
    return int(result)

def compare_euler_phi_methods(test_values: List[int]) -> Dict[str, List[float]]:
    """Сравнивает время работы трёх методов на заданных значениях, возвращает словарь с тремя списками времён (в секундах)."""
    times_direct = []
    times_factor = []
    times_sympy = []
    for n in test_values:
        start = time.time()
        euler_phi_direct(n)
        times_direct.append(time.time() - start)

        start = time.time()
        euler_phi_factor(n)
        times_factor.append(time.time() - start)

        start = time.time()
        totient(n)
        times_sympy.append(time.time() - start)

    return {
        'direct': times_direct,
        'factorization': times_factor,
        'sympy': times_sympy
    }

result_6 = compare_euler_phi_methods([10**3, 10**4, 10**5])
```

```python
#задание 1 с ответом

from sympy.combinatorics.named_groups import SymmetricGroup, AlternatingGroup
from sympy.combinatorics.permutations import Permutation
from sympy.combinatorics.perm_groups import PermutationGroup
import random

def get_cosets(G, H, side='left'):
    reps = []
    elements_G = list(G.generate())
    elements_H = set(H.generate())
    marked = set()
    for g in elements_G:
        if g in marked:
            continue
        if side == 'left':
            coset = set([g * h for h in elements_H])
        else:
            coset = set([h * g for h in elements_H])
        reps.append(list(coset))
        marked.update(coset)
    return reps

def subgroups_of_Sm(N, m):
    Sm = SymmetricGroup(m)
    subgroups = []
    subgroups.append(PermutationGroup([Sm.identity]))
    subgroups.append(Sm)
    if m >= 3:
        subgroups.append(AlternatingGroup(m))
    cycle = Permutation(list(range(m)))
    subgroups.append(PermutationGroup([cycle]))
    fix0 = [p for p in Sm.generate() if p(0) == 0]
    subgroups.append(PermutationGroup(fix0))
    total = len(subgroups)
    random_subgroup = random.choice(subgroups)
    N_mod = N % total
    S = subgroups[N_mod]
    index = Sm.order() // S.order()
    is_normal = S.is_normal(Sm)
    left_cosets = get_cosets(Sm, S, 'left')
    right_cosets = get_cosets(Sm, S, 'right')
    return {
        "subgroups_count": total,
        "random_subgroup": random_subgroup,
        "target_subgroup": S,
        "index": index,
        "is_normal": is_normal,
        "left_cosets": left_cosets,
        "right_cosets": right_cosets,
    }

result = subgroups_of_Sm(N=3, m=7)
#ответ
print(result)
```

```python
from sympy import symbols, Poly, degree
from itertools import product

def is_irreducible_custom(poly, q):
    x = poly.gens[0]
    d = poly.degree()
    for deg in range(1, d // 2 + 1):
        for coeffs in product(range(q), repeat=deg):
            if coeffs[0] == 0:
                continue
            divisor = Poly(sum(coeffs[i] * x**(deg-1-i) for i in range(deg)) + x**deg, x, modulus=q)
            rem = poly.div(divisor)[1]
            if rem.is_zero:
                return False
    return True

def generate_irreducible_polynomials(q: int, d: int) -> list:
    x = symbols('x')
    polys = []
    for coeffs in product(range(q), repeat=d):
        if coeffs[0] == 0:
            continue
        poly = Poly(sum(coeffs[i] * x**(d-1-i) for i in range(d)) + x**d, x, modulus=q)
        if is_irreducible_custom(poly, q):
            polys.append(poly)
    return polys

result = generate_irreducible_polynomials(2, 3)
for p in result:
    print(p)
```
